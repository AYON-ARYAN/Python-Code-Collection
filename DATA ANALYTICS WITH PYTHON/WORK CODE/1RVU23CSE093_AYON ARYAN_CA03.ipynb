```python
import pandas as pd
import matplotlib.pyplot as plt


data = {
    'Food': ['100% Bran', '100% Natural Bran', 'All-Bran', 'All-Bran with Extra Fiber', 'Almond Delight'],
    'calories': [70, 120, 70, 50, 110],
    'protein': [4, 3, 4, 4, 2],
    'fat': [1, 5, 1, 0, 2],
    'sodium': [130, 15, 260, 140, 200],
    'fiber': [10, 2, 9, 14, 1],
    'carbo': [5, 8, 7, 8, 14],
    'sugars': [6, 8, 5, 0, 8],
    'potass': [280, 135, 320, 330, -1],
    'vitamins': [25, 0, 25, 25, 25]
}

df = pd.DataFrame(data)


x_column = 'calories'
y_column = 'protein'


plt.figure(figsize=(10, 6))
plt.scatter(df[x_column], df[y_column], c='blue', marker='o')


plt.xlabel(x_column.capitalize())  
plt.ylabel(y_column.capitalize())
plt.title(f'Scatter Plot of {x_column.capitalize()} vs {y_column.capitalize()}')


for i, food in enumerate(df['Food']):
    plt.annotate(food, (df[x_column][i], df[y_column][i]), textcoords="offset points", xytext=(5, 5), ha='right')


plt.show()
```
