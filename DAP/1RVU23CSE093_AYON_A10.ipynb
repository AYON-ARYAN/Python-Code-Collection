```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.title("Graph of years of experiance")
plt.xlabel("Years of experiance")#name of x axis
plt.ylabel("frequency")#name of y axis
sns.histplot(a['YearsExperience'],color= 'red')#change color
```

```python
plt.title("Graph of years of experiance")
plt.xlabel("Years of experiance")#name of x axis
plt.ylabel("frequency")#name of y axis
sns.histplot(a['YearsExperience'],color= 'silver', bins =80)#change color<bins increase the number of bars
```

```python
plt.title("Graph of salary")
plt.xlabel("Salary")#name of x axis
plt.ylabel("Number of people")#name of y axis
sns.histplot(a['Salary'],color= 'silver', bins= 111)
```
