**Source of Dataset:** Data were collected and made available by Dr. Kristen Gorman and the Palmer Station, Antarctica LTER, a member of the Long Term Ecological Research Network.

### **Details of the Dataset**
**penguins_size.csv:** Simplified data from original penguin data sets. Contains variables:

1. **species:** penguin species (Chinstrap, Ad√©lie, or Gentoo)
2. **culmen_length_mm:** culmen length (mm)
3. **culmen_depth_mm:** culmen depth (mm)
4. **flipper_length_mm:** flipper length (mm)
5. **body_mass_g:** body mass (g)
6. **island:** island name (Dream, Torgersen, or Biscoe) in the Palmer Archipelago (Antarctica)
7. **sex:** penguin gender

```python
#Import the neccessary library for for the assigned task
#Write your code below this line
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#pre-code to load the file 
df = pd.read_csv('penguins.csv')
df.head(3)
```

### Plots 

```python
#Create the distribution plot for penguin species 
#Write your code below this line

sns.displot(df['species'], kde=True, color='blue', bins=30)

plt.xlabel('Values')
plt.ylabel('Density')
plt.title(f'Distribution Plot of penguin species')
plt.show()
```

```python
#Create the scatter plot for x = body_mass_g and y = flipper_length_mm of penguins.
#Scenario - study of relationship below penguin flipper length and penguin body mass.





sns.scatterplot(x='body_mass_g', y='flipper_length_mm', data=df)


plt.xlabel('Body Mass (g)')
plt.ylabel('Flipper Length (mm)')
plt.title('Scatter Plot of Penguin Flipper Length vs Body Mass')


plt.show()
```

#### [edit this markdown cell and mention your answer]

**Question 1:** Which Bird have species high body mass ?

**Answer:** gentoo

```python
#Create the scatter plot for x = bill_depth_mm and y = bill_length_mm of penguins and categorize in terms of species
#Hint - Pass the {HUE = Species} parameter in Scatter plot method.
#Scenario - study of relationship below bill_depth_mm and bill_length_mm for different species.

#Write your code below this line
sns.scatterplot(x='bill_depth_mm', y='bill_length_mm', hue='species', data=df)


plt.xlabel('Bill Depth (mm)')
plt.ylabel('Bill Length (mm)')
plt.title('Scatter Plot of Penguin Bill Length vs Bill Depth (Categorized by Species)')


plt.legend()


plt.show()
```

```python
#Create the box plot for penguin 
#Scenario - study of outliers for the penguins flipper_length_mm .

#Write your code below this line
sns.boxplot(x='flipper_length_mm', y='species', data=df)


plt.xlabel('Flipper Length (mm)')
plt.ylabel('Species')
plt.title('Box Plot of Penguin Flipper Lengths (Study of Outliers)')


plt.show()
```
