```python
#Write your code below this line.
df7=pd.read_csv("insurance.csv")
df7
```

```python
#plot the bar chart for the above dataset

#1 Create a figure object
figure = plt.figure(figsize=(15,10))

#2 Create a bar chart
#plt.plot(df4['age'], df4['charges'],color = 'blue', linewidth = 0.1, alpha = 0.3)
sns.lineplot(data = df7, x= 'age', y = 'charges',color = 'blue',marker='o' ,hue = 'smoker')
# Make the background grid less distracting
plt.grid(alpha=0.25)

#3 Add labels and titles to the chart
plt.xlabel('Age')
plt.ylabel('Charges of Insurance Premium')
plt.title("Age VS Insurance Charges")

#4 Display the chart
plt.show()
```

**Write the inference w.r.t above Chart** 

#Write your answer here
The Insurance Cover of the People who Smoke is Higher when compared to others.
