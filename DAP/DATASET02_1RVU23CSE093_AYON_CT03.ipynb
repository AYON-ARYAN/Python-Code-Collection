```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("/Users/ayon_aryan/Downloads/Dataset02.csv")
df
```

```python
df.info()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='units_sold', y='price', hue='category', data=df, palette='viridis', alpha=0.7)
plt.title(' Units Sold vs Price')
plt.xlabel('Units Sold')
plt.ylabel('Price')
plt.legend(title='Category')  
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='customer_reviews', data=df, palette='deep')
plt.title(' Customer Reviews by Category')
plt.xlabel('Category')
plt.ylabel('Customer Reviews')
plt.show()
```

```python

plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='units_sold', data=df, ci=None, palette='viridis')
plt.title('Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='category', y='customer_reviews', data=df, ci=None, palette='viridis')
plt.title('Customer Reviews by Category and Rating')
plt.xlabel('Category')
plt.ylabel('Customer Reviews')
plt.legend(title='Rating')
plt.show()
```
