```python
#Import the neccessary library for for the assigned task
#Write your code below this line
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#load the file and display the first five rows
#Write your code below this line
df=pd.read_csv("/Volumes/PEN DRIVE/DATA ANALYTICS WITH PYTHON/flights.csv")
df.head(5)
```

### Plots

```python
#Create the line plot with proper labels as decided by self for x = Years and y = Passengers.
#Scenario - study of trend of Airline travel.

#Write your code below this line


# Read data from CSV file



years = df['year']
passengers = df['passengers']


plt.plot(years, passengers, marker='o', linestyle='-')


plt.xlabel('Years')
plt.ylabel('Passengers')
plt.title('Airline Travel Trend Over the Years')


plt.grid(True)  
plt.show()
```

```python
#Create the bar plot for x = months and y = Passengers.
#Scenario - Study of distribution of monthly wise passengers travels.
#Hint: plt.bar(df['column1'], df['column2'])
#Write your code below this line


plt.bar(df['month'], df['passengers'], color='red', edgecolor='black')


plt.xticks(rotation=45)
plt.xlabel('Months')
plt.ylabel('Passengers')
plt.title('Distribution of Monthly Wise Passengers Travel')


plt.grid(axis='y')  
plt.show()
```

```python
#Create the histogram plot.
#Scenario - study of passenger distribution in the dataset.

#Write your code below this line



plt.hist(df['passengers'], bins=10, color='black', edgecolor='red')


plt.xlabel('Passengers')
plt.ylabel('Frequency')
plt.title('Passenger Distribution in the Dataset')


plt.grid(axis='y')  
plt.show()
```

# SELF EXPLORATORY TASK [ INTERMEDIATE LEVEL ]

**Study of Detailed Monthly wise Passengers travelling information**

```python
df1 = df.pivot(index="year", columns="month", values="passengers")
df1.head()
```

```python
#Create the line plot with with parameter data = df1



df1.plot(marker='o', linestyle='-', figsize=(10, 6))


plt.xlabel('Year')
plt.ylabel('Passengers')
plt.title('Passenger Trend Over Months and Years')


plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')


plt.grid(True)

plt.show()
```
